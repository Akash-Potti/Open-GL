```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
# Read the image
plt.figure(figsize=(20,10))
rows=2
col=2
src = cv2.imread('img.png')
plt.subplot(rows,col,1)
plt.imshow(src)
plt.axis('off')
plt.title("original")

rotate=cv2.rotate(src,cv2.ROTATE_90_CLOCKWISE)
plt.subplot(rows,col,2)
plt.imshow(rotate)
plt.axis('off')
plt.title("Rotated")
(h,w)=src.shape[:2]
newdim=(100,h)
resize=cv2.resize(src,newdim,interpolation=cv2.INTER_AREA)
plt.subplot(rows,col,3)
plt.imshow(resize)
plt.axis('off')
plt.title("Resize")
m = np.float32([[1, 0, 25], [0, 1, 50]])  
shift = cv2.warpAffine(src, m, (w, h))
plt.subplot(rows, col, 4)
plt.imshow(shift)
plt.axis('off')
plt.title("Translated")
```
