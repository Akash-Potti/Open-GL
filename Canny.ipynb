```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('img.png')

grey=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edges=cv2.Canny(grey,100,200)
blurred = cv2.blur(img, (50, 50))
smoother=cv2.GaussianBlur(img,(51,51),0)
plt.figure(figsize=(7,8))
plt.subplot(2,4,1)
plt.imshow(img)
plt.subplot(2,4,2)
plt.imshow(edges)
plt.subplot(2,4,3)
plt.imshow(grey)
plt.subplot(2,4,4)
plt.imshow(blurred)
plt.subplot(2,4,5)
plt.imshow(smoother)
plt.show()
plt.axis('off')
```
