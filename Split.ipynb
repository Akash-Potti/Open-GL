```python
import cv2
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('img.png')

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(7, 8))
row = 3
columns = 2

(h, w) = image.shape[:2]

(cx, cy) = (w // 2, h // 2)

topleft = image[0:cy, 0:cx]
topright = image[0:cy, cx:w]
bottomleft = image[cy:h, 0:cx]
bottomright = image[cy:h, cx:w]

plt.subplot(row, columns, 1)
plt.imshow(image)
plt.axis('off')
plt.title('Original')

plt.subplot(row, columns, 3)
plt.imshow(topleft)
plt.axis('off')
plt.title('Top Left')

plt.subplot(row, columns, 4)
plt.imshow(topright)
plt.axis('off')
plt.title('Top Right')

plt.subplot(row, columns, 5)
plt.imshow(bottomleft)
plt.axis('off')
plt.title('Bottom Left')

plt.subplot(row, columns, 6)
plt.imshow(bottomright)
plt.axis('off')
plt.title('Bottom Right')

# Show the plot
plt.tight_layout()
plt.show()
```
